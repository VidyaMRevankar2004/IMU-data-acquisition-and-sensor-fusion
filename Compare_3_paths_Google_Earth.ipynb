```python
import pandas as pd
import simplekml

# File paths
kf_file = '/home/pi/Desktop/KF_Filtered_Output.csv'  # Kalman Filter output
gt_file = '/home/pi/Desktop/20250427-115044.txt'    # Ground Truth
sensor_file = '/home/pi/Desktop/imu_gps_log_20250427_115034.csv'  # Original sensor data

# Read the data
kf_data = pd.read_csv(kf_file)
gt_data = pd.read_csv(gt_file)
sensor_data = pd.read_csv(sensor_file)

# Strip any extra spaces from column names
kf_data.columns = kf_data.columns.str.strip()
gt_data.columns = gt_data.columns.str.strip()
sensor_data.columns = sensor_data.columns.str.strip()

# Create KML object
kml = simplekml.Kml()

# Add Ground Truth path
gt_coords = list(zip(gt_data['longitude'], gt_data['latitude']))  # Ensure correct case & strip spaces
gt_linestring = kml.newlinestring(name="Ground Truth Path")
gt_linestring.coords = gt_coords
gt_linestring.style.linestyle.color = simplekml.Color.red
gt_linestring.style.linestyle.width = 4

# Add Kalman Filter path
kf_coords = list(zip(kf_data['Filtered_Latitude'], kf_data['Filtered_Longitude']))  # Filtered lat/lon
kf_linestring = kml.newlinestring(name="Kalman Filtered Path")
kf_linestring.coords = kf_coords
kf_linestring.style.linestyle.color = simplekml.Color.green
kf_linestring.style.linestyle.width = 4

# Add Original Sensor Data path
sensor_coords = list(zip(sensor_data['Lon'], sensor_data['Lat']))  # Sensor lat/lon
sensor_linestring = kml.newlinestring(name="Original Sensor Path")
sensor_linestring.coords = sensor_coords
sensor_linestring.style.linestyle.color = simplekml.Color.blue
sensor_linestring.style.linestyle.width = 4

# Save KML file
output_kml = '/home/pi/Desktop/compare_three_paths.kml'
kml.save(output_kml)

print(f"KML file saved at {output_kml}")
```
